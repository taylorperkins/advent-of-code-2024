```python
from itertools import cycle
import pickle

import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

plt.rcParams["animation.html"] = "jshtml"
plt.style.use('ggplot')

import numpy as np

%matplotlib inline
```

```python
with open("out.pkl", "rb") as f:
    maps = pickle.load(f)

keys = sorted(maps.keys())
```

```python
fig, ax = plt.subplots()

cax = ax.pcolormesh(maps[keys[0]], cmap='Blues')
fig.colorbar(cax)

def animate(i):
    if keys[i] > 7753:
        ax.set_title(7753)
        cax.set_array(maps[7753][::-1])
    else:
        ax.set_title(keys[i])
        cax.set_array(maps[keys[i]][::-1])

ani = FuncAnimation(
    fig, animate, interval=100, frames=len(keys))
        
from IPython.display import HTML
HTML(ani.to_jshtml())

ani.save("part2_animation.mp4")
```
